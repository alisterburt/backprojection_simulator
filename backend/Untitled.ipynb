```python
from backprojection import get_single_projection, ramp_filter_500, fft_projection, apply_ramp
from skimage.io import imread, imsave
from skimage.transform import rescale, rotate
from scipy.fft import ifft
import matplotlib.pyplot as plt
from image import grayscale, square, preprocess
```

```python
im = preprocess('rabbit.png')
im.shape
```

```python
proj = get_single_projection(im, 90)
```

```python
proj.shape
```

```python
fig, ax = plt.subplots(2,1)
ax[0].imshow(rotate(im,90), cmap='gray_r')
ax[1].plot(proj)
```

```python
ft = fft_projection(proj)
```

```python
ft.shape
```

```python
ramp = ramp_filter_500()
```

```python
ft = ft * ramp
```

```python
fig, ax = plt.subplots()
ax.plot(ft)
```
